```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("G:/Shared drives/echolab data/wildfire_speciation/intermediate/pm_plume_speciation_at_sites.csv")
first_column = df.columns[0]
df = df.drop([first_column], axis=1)
```

```python
# drop State == NaN, Region == Nan
df_clean = df.dropna(subset=['State', 'region'])
df_clean.isna().sum()
```

```python
df_clean['region'].unique()
```

```python
# keep relevant columns only
df_sum = df_clean[['smokePM_pred', 'MF', 'RC_PM2', 'NH4', 
           'tot_metals', 'ammNO3', 'ammSO4', 'AL', 'AS','BR', 'CA', 'EC', 
           'OC', 'CHL', 'CL', 'CR', 'CU', 'FE', 'K', 'MG', 'MN', 'MO', 'NA', 
           'NI', 'NO3', 'N2', 'P', 'PB', 'RB', 'S', 'SE', 'SI', 'SOIL','SO4', 
           'SR', 'V', 'ZN', 'ZR']]
```

```python
pd.set_option('display.max_columns',None)
round(df_sum.describe(),2)
```

```python
# drop inifinite values from NH4 
df_filter = df.isin([np.inf, -np.inf])
df_sum_clean = df_sum[~df_filter]
```
